3. *Limpieza y conversión de campos*

4. *Normalización y verificación final*


• Verifique valores nulos y duros:

```python
df.isnull().sum()
```

```python
df["Weight"]=df["Weight"].fillna(df["Weight"].median())
df["Type2"]=df["Type2"].fillna("None")
df["Base Exp"]=df["Base Exp"].fillna(df["Base Exp"].median())
df["Female (%)"]=df["Female (%)"].fillna("0")
```

Valores duros

```python
def valores_duros(variable) :
    umbral = 1.5
    q1 = variable.quantile(0.25)
    q3 = variable.quantile(0.75)
    iqr = q3 - q1
    limite_inferior = q1 - umbral * iqr
    limite_superior = q3 + umbral * iqr
    valores = variable[(variable < limite_inferior) | (variable > limite_superior)]
    return valores

valores_duros(df['Weight'])
valores_duros(df['Height'])
valores_duros(df['HP Base'])
valores_duros(df['Attack Base'])
valores_duros(df['Defense Base'])
valores_duros(df['Speed Base'])
valores_duros(df['Base Exp'])
```

Determine el promedio de altura y peso por tipo principal (Type1).
